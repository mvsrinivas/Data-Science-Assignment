```python
from sklearn.cluster import KMeans

# Choose the number of clusters (e.g., 4)
kmeans = KMeans(n_clusters=4, random_state=42)
customer_profiles['Cluster'] = kmeans.fit_predict(customer_profiles[['TotalSpending', 'NumTransactions']])
```

```python
from sklearn.metrics import davies_bouldin_score

db_index = davies_bouldin_score(customer_profiles[['TotalSpending', 'NumTransactions']], customer_profiles['Cluster'])
print(f"Davies-Bouldin Index: {db_index}")
```

```python
sns.scatterplot(x='TotalSpending', y='NumTransactions', hue='Cluster', data=customer_profiles, palette='viridis')
plt.title('Customer Clusters')
plt.show()
```
